```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/archive.zip")
df.head()
```

# Choose coords for an east and a west boundary as corners of a rectangle ðŸ—º

create custom map  https://www.google.com/maps/d/

```python
# lat, lon
WEST = (32.83157, -117.22127)
EAST = (32.77472, -117.16325)

# Write a functiom for location within bounds of east and west
def locate_me(lat, lon):
    w = WEST
    e = EAST
    if lat < w[0] and lon > w[1] and lat > e[0] and lon < e[1]:
        return True
    else:
        return False
```

# Create new feature

```python
# use vectorize to apply function to different columns
df['clairemont'] = np.vectorize(locate_me)(df['latitude'], df['longitude'])
```
